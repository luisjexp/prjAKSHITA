```python
def vis_race_tt_Xcont_Yevent(tt_race, Xcont_name, Yevent_name):
  timestamps  = tt_race['Time']
  Xcont_vct    = tt_race[Xcont_name] 

  Yevent_list_all  = tt_race[Yevent_name]
  Yevent_idx    = Yevent_list_all.notnull()
  Yevent_list = Yevent_list_all[Yevent_idx]
  Yevent_times = timestamps[Yevent_idx]
   
  Yevent_lineheight_unit_ax = np.linspace(.1, .9 , len(Yevent_times))
  Yevent_lineheight_unit_data = np.linspace(Xcont_vct.min(), Xcont_vct.max() , len(Yevent_times))


  ax = plt.subplots(1,1, figsize=(100, 10))
  plt.plot(timestamps, Xcont_vct, color ='tab:blue') 
  # plt.show()
  for count, t in enumerate(Yevent_times):
    lineheight_unit_ax = Yevent_lineheight_unit_ax[count]
    lineheight_unit_data = Yevent_lineheight_unit_data[count]
    Yevent_describe = Yevent_list.iloc[count]
    print(t, Yevent_describe)

    plt.axvline(x = t, ymin=0, ymax=lineheight_unit_ax )
    plt.text(t, lineheight_unit_data, Yevent_describe, fontdict=None, horizontalalignment='right',
             verticalalignment='center', rotation=80)
    
    
    
  plt.xlim(min(Yevent_times)  , max(Yevent_times)) 
 
    # plt.show()
    # print(lineheight)


  # plt.vlines(Yevent_times, Xcont_vct.min(), Xcont_vct.max())
  # plt.title('matplotlib.pyplot.plot() example 2') 
  # plt.show()

 


## RUN
# print(tt_race.columns)

tt_race     = tt_wht
Xcont_name  = 'calgov_case_count_daily'
Yevent_name = 'aks_npi_description'
vis_race_tt_Xcont_Yevent(tt_race, Xcont_name, Yevent_name)
 
```
